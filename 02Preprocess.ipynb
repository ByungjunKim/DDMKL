```
# 일반/고유명사만 추출
def khaii_tokenize(sent):
    token = []
    for word in api.analyze(sent):
        for morph in word.morphs:
            if morph.tag in ['NNG','NNP']: 
                token.append(morph.lex)
    return token
```

```
# 일반/고유명사, +접미사, + 접두어, 숫자+의존명사
def khaii_tokenize2(sent):
    token = []
    for word in api.analyze(sent):
        if len(word.morphs)==1:
            for morph in word.morphs:
                if morph.tag in ['NNG','NNP']: # 일반/고유명사만 추출
                    token.append(morph.lex)
        else:
            comp_tags = '+'.join([w.tag for w in word.morphs])
            if bool(re.match('SN\+NN[BG]',comp_tags)) or bool(re.match('XPN\+NN[PG]',comp_tags) or bool(re.match('NN[PG]\+XSN',comp_tags))):
                if bool(re.match('NN[PG]\+XSN',comp_tags)) and ([w.lex for w in word.morphs][1]=='적'):
                    token.append(''.join([w.lex for w in word.morphs][0:1])) # 일반/고유명사 + 명사 접미사 일때 접미사가 '적'이라면 앞에 명사만 추출
                else:
                    token.append(''.join([w.lex for w in word.morphs][0:2])) # 숫자+의존명사, 접두어 +일반/고유명사, 일반/고유명사 + 명사 접미사 
            else:
                for morph in word.morphs:
                    if morph.tag in ['NNG','NNP']: # 일반/고유명사만 추출
                        token.append(morph.lex)
    return token
```

```
# 동의어 처리(시대/연도 등)
def multipleReplace(text, wordDict):
    for key in wordDict:
        # text = text.replace(key, wordDict[key])
        text = re.sub(f'^{key}$', wordDict[key],text)
    return text

# 동의어 사전
# origin word : new word
synonym_dict = {
    '10년대' : '1910년대',
    '20년대' : '1920년대',
    '30년대' : '1930년대',
    '40년대' : '1940년대',
    '50년대' : '1950년대',
    '60년대' : '1960년대',
    '70년대' : '1970년대',
    '80년대' : '1980년대',
    '90년대' : '1990년대',
    '00년대' : '2000년대',
    '1900' : '1900년대',
    '1910' : '1910년대',
    '1920' : '1920년대',
    '1930' : '1930년대',
    '1940' : '1940년대',
    '1950년' : '1950',
    '1960' : '1960년대',
    '1970' : '1970년대',
    '1980' : '1980년대',
    '1990' : '1990년대',
    '2000' : '2000년대',
    '데올로기':'이데올로기',
}
```

```
# 동의어 처리 실행
# phd['tokens'] = phd['tokens'].map(lambda x:multipleReplace(','.join(x),synonym_dict).split(','))
phd['tokens'] = phd['tokens'].progress_map(lambda x:[multipleReplace(w,synonym_dict) for w in x])
```

```
# 1음절 제거 (유효한 1음절은 살리기)
hangul_1_except = regex.compile(r'^(?!삶|시|곡|애|정|힘|글|미|선|덕|체|흙|심|책|앎|죄|꿈|웹|몸|틀|빛|말|피|벽|극)\p{Hangul}{1}$')
phd['tokens'] = phd['tokens'].progress_map(lambda x:[w for w in x if not hangul_1_except.match(w)]) #매치되는 것을 제거한다.
```
